## Basic plotting

```python
import mplfinance as mpf
```

```python
aapl = yf.Ticker('aapl')
mpf.plot(
    aapl.history(period='2y', interval='5d'),
    volume=True,
    mav=(3,6,9),
    type='candle',
    style='yahoo',
)
```
